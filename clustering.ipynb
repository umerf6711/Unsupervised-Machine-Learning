# 1. Affinity propagation

```python
from numpy import where
from numpy import unique
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import AffinityPropagation
```

```python
X, _ = make_classification(n_samples=1000,n_features=2,n_informative=2,n_redundant=0,n_clusters_per_class=1,random_state=4)
model = AffinityPropagation(damping=0.9)
model.fit(X)

```

```python
yhat=model.predict(X)
clusters = unique(yhat)
```

```python
for cluster in clusters:
    row_ix = where(yhat==cluster)
    plt.scatter(X[row_ix,0],X[row_ix,1])
plt.show()    
```

# 2.Agglomerative

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model1= AgglomerativeClustering(n_clusters=2)
```

```python
yhat = model1.fit_predict(X)
clusters = unique(yhat)
```

```python
for cluster in clusters:
    row_ix = where(yhat==cluster)
    plt.scatter(X[row_ix,0],X[row_ix,1])
plt.show() 
```

# 3. Birch

```python
from sklearn.cluster import Birch
```

```python
model2 = Birch(threshold=0.01,n_clusters=2)
model2.fit(X)
yhat = model2.predict(X)
clusters  = unique(yhat)
```

```python
for cluster in clusters:
    row_ix = where(yhat==cluster)
    plt.scatter(X[row_ix,0],X[row_ix,1])
plt.show() 
```

# 4. DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
model3 = DBSCAN(eps=0.30,min_samples=9)
model3.fit(X)
yhat = model3.fit_predict(X)
clusters = unique(yhat)
```

```python
for cluster in clusters:
    row_ix = where(yhat==cluster)
    plt.scatter(X[row_ix,0],X[row_ix,1])
plt.show() 
```

# 5. KMeans

```python
from sklearn.cluster import KMeans
```

```python
model4 = KMeans(n_clusters=2)
yhat = model4.fit_predict(X)
clusters = unique(yhat)
```

```python
for cluster in clusters:
    row_ix = where(yhat==cluster)
    plt.scatter(X[row_ix,0],X[row_ix,1])
plt.show() 
```
