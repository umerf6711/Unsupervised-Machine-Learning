# Silhouette coefficient

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
X = np.random.rand(50,2)
Y = 2+ np.random.rand(50,2)
Z = np.concatenate((X,Y)
Z = pd.DataFrame(Z)
```

```python
sns.scatterplot(Z[0],Z[1])
```

```python
model = KMeans(n_clusters=2)
```

```python
model.fit(Z)
```

```python
label = model.predict(Z)
```

```python
print('silhouette score(n=2):',silhouette_score(Z,label))
```

```python
#plot with labeled data points

sns.scatterplot(Z[0],Z[1],hue=label)
```

```python
#for 3 clusters 

model = KMeans(n_clusters=3)
model.fit(Z)
label=model.predict(Z)
print('silhouette score(n=3):',silhouette_score(Z,label))
sns.scatterplot(Z[0],Z[1],hue=label)
```

**Result**

*We may notics n=2 is more efficient and we may use silhouette to find the number of clusters.*

```python
from sklearn import datasets
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
from sklearn.datasets import make_blobs
```

```python
X ,y_true = make_blobs(n_samples=300,centers=4,cluster_std=0.50,random_state=0)
kmean = KMeans(n_clusters=4,random_state=1).fit(X)
labels = kmean.labels_
```

```python
print(davies_bouldin_score(X,labels))
```
